```
import pandas as pd
import numpy as np


df=pd.read_csv('datasetlab4.txt',header=None)
df=df.replace("?","abcd")

df
```

```
#df.iloc[1:1,0:6]
dim=df.shape
X=dim[0]
Y=dim[1]

for v in range(128):
  if(v!=3):
    c=0
    n=0
    arr=df.iloc[0:X,v:v+1]
    #print(arr)
    arr=arr.to_numpy()
    #print(arr)
    for i in range (X):
      if(arr[i][0]!="abcd"):
        c=c+float(arr[i][0])
        n=n+1
    print(c/n)
    df.iloc[0:X,v:v+1]=df.iloc[0:X,v:v+1].replace("abcd",str(c/n))    
df        
```

```
coliarr=[]
col3arr= df.iloc[0:X,3:4].to_numpy()
for i in range(X):
  if(col3arr[i][0] not in coliarr):
    coliarr.append(col3arr[i][0])
print(len(coliarr))
# for i in range(len(coliarr)):
#   if(coliarr[i]=="Columbuscity"):
#     print(i,"************")
```

```
for i in range(len(coliarr)):
   df.insert(127+i,coliarr[i], 0)
print(df.shape) 
df
```

```
for i in range(1890,1994):
    s=df.iloc[i:i+1,3:4].to_numpy()
    df.loc[i,str(s[0][0])]=1
df

# #df["Walthamcity"]=1
# #df.loc[5,"Walthamcity"]=100
# df
```

```
for ii in range(0,5):
  if(ii!=3):
   narr=df.iloc[0:X,ii:ii+1].to_numpy()
   maxm=0
   minm=100000
   for i in range(X):
     if(maxm<float(narr[i][0])):
        maxm=float(narr[i][0])
     if(minm>float(narr[i][0])):
        minm=float(narr[i][0])
   for i in range(X):
        df.loc[i,ii]=str((float(df.loc[i,ii])-minm)/(maxm-minm))
del df[3]
df
```

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
lin_reg = LinearRegression()
lin_reg.fit(df.iloc[:1500,:-1],df.iloc[:1500,1954:] )
y_pred_0 = lin_reg.predict(df.iloc[1500:,:-1])
print("Training error (RMSE) :", mean_squared_error(df.iloc[:1500,1954:], lin_reg.predict(df.iloc[:1500,:-1])))
print("Testing error (RMSE) :", mean_squared_error(df.iloc[1500:,1954:], y_pred_0))
```
